```python
def create_sequences(data, input_tw, output_tw, interval):
    inout_seq = []
    L = len(data)
    for i in range(0, (L-input_tw), interval):
        _seq = data[i:i+input_tw]
        _label = data[i+input_tw:i+input_tw+output_tw]
        inout_seq.append((_seq ,_label))
    return inout_seq
```

```python
seq_out = create_sequences(all_stock_data_AAL, 84, 14, 28)
```
